# 游닌 Download Samples

At present, the most popular websites to download field recordings are [Xeno Canto](https://xeno-canto.org/) and [eBird](https://ebird.org/home). The main difference is the data adquisition. To download data from eBird you have to request the audios, [Requesting and downloading media](https://support.ebird.org/en/support/solutions/articles/48001064551-requesting-and-downloading-media), while the datos in Xeno Canto are public and can be donwload via the website or via API, [Application Programming Interface (API v2)](https://xeno-canto.org/explore/api).


## 游꿨 Xeno Canto

You can download audio files from Xeno-Canto using the `wavesongs.data.xc` module. First, create a query for the desired sounds. To do this, you need the scientific name of the specie. You can also include additional details such as the country or location where the recording was made to refine your search. Check the [Xeno Canto API](https://xeno-canto.org/explore/api) to check other possible filters.

You can also query multiple species by providing a list of common names along with their corresponding scientific names.

```python
english_name = ["Rufous-collared Sparrow", "Eurasian Three-toed"]
scientific_name = ["Zonotrichia capensis", "Picoides tridactylus"]
```

### Query Audios

````python
from wavesongs.data.xc import download_audios, query_audios

info = {
  "cnt": "Colombia",
  "loc": "Bogota",
}

df_query = query_audios(
    english_name = "Rufous-collared Sparrow",
    scientific_name = "Zonotrichia capensis",
    info = info,
    max_nb_files=None, # Number of files to download
    format_time=True,
    format_date=True  
)

df_query
````

### Download Audios

Define the download directory (`rootdir`) for the audio files selected in the query using the dataset parameters. If no dataset name is specified, the default name will be the scientific name, with spaces replaced by underscores (_).

````python
# Download the first 5 files
df_query_short = df_query.iloc[:5]

df_audios = download_audios(
    df_query_short,
    rootdir = "./assets/audios", 
    dataset_name = '',
    overwrite = True,
    save_csv = True,
    verbose = True
)

df_audios
````
